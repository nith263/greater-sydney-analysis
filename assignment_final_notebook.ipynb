```python
from sqlalchemy import create_engine
import psycopg2
import psycopg2.extras
import json
import os
import pandas as pd

credentials = "Credentials.json"

def pgconnect(credential_filepath, db_schema="public"):
    with open(credential_filepath) as f:
        db_conn_dict = json.load(f)
        host       = db_conn_dict['host']
        db_user    = db_conn_dict['user']
        db_pw      = db_conn_dict['password']
        default_db = db_conn_dict['user']
        try:
            db = create_engine('postgresql+psycopg2://'+db_user+':'+db_pw+'@'+host+'/'+default_db, echo=False)
            conn = db.connect()
            print('Connected successfully.')
        except Exception as e:
            print("Unable to connect to the database.")
            print(e)
            db, conn = None, None
        return db,conn
    
def query(conn, sqlcmd, args=None, df=True):
    result = pd.DataFrame() if df else None
    try:
        if df:
            result = pd.read_sql_query(sqlcmd, conn, params=args)
        else:
            result = conn.execute(sqlcmd, args).fetchall()
            result = result[0] if len(result) == 1 else result
    except Exception as e:
        print("Error encountered: ", e, sep='\n')
    return result
```

```python
db, conn = pgconnect(credentials)
```

```python
srid = 4326
def create_wkt_element(geom, srid):
    if geom is None:
        return None
    if geom.geom_type == 'Polygon':
        geom = MultiPolygon([geom])
    return WKTElement(geom.wkt, srid)
```

## 2.4 Joining Tables

```python
sql = """
DROP TABLE IF EXISTS BusinessPopulation;
CREATE TABLE BusinessPopulation AS
SELECT R.sa2_code, R.sa2_name, R.r_per_1000, H.h_per_1000, S.geom
FROM RetailTrade AS R
JOIN HealthCare AS H ON R.sa2_code = H.sa2_code
JOIN Population AS P ON R.sa2_code = P.sa2_code
JOIN SA2 AS S ON R.sa2_code = S.sa2_code
"""

query(conn, sql)
query(conn, "select COUNT(*) from BusinessPopulation")
```

```python
sql = """
DROP TABLE IF EXISTS PSchools;
CREATE TABLE PSchools AS
SELECT BP.*, COUNT(*) AS Primary_Schools
FROM BusinessPopulation AS BP
LEFT JOIN PrimaryCatch AS Pr
ON ST_Contains(BP.geom, Pr.geom)
GROUP BY BP.sa2_code, BP.sa2_name, BP.r_per_1000, BP.h_per_1000, BP.geom
"""
query(conn, sql)
```

```python
query(conn, "select * from pschools")
```

```python
sql = """
DROP TABLE IF EXISTS PSSchools;
CREATE TABLE PSSchools AS
SELECT PS.*, COUNT(*) AS secondary_schools
FROM PSchools AS PS
LEFT JOIN Secondary AS S
ON ST_Intersects(PS.geom, S.geom)
GROUP BY PS.sa2_code, PS.sa2_name, PS.r_per_1000, PS.h_per_1000, PS.geom,  PS.primary_schools
"""
query(conn, sql)
```

## 3.1 Compute Z-Scores

```python
sql = """
DROP TABLE IF EXISTS ZScores;
CREATE TABLE ZScores(
    sa2_code int,
    sa2_name VARCHAR(50),
    retail_zs float,
    health_zs float,
    primary_zs float,
    secondary_zs float,
    future_zs float,
    stops_zs float,
    polling_zs float,
    geom GEOMETRY(MULTIPOLYGON, 4326)
)
"""
query(conn, sql)
```

```python
sql = """
WITH stats AS(
    SELECT AVG(r_per_1000) AS avg_retail,
           STDDEV(r_per_1000) AS stddev_retail,
           AVG(h_per_1000) AS avg_health,
           STDDEV(h_per_1000) AS stddev_health,
           AVG(primary_schools) as avg_prim,
           STDDEV(primary_schools) as stddev_prim,
           AVG(secondary_schools) as avg_second,
           STDDEV(secondary_schools) as stddev_second,
           AVG(tansport_count) as av_stops,
           STDDEV(tansport_count) as stddev_stops,
           AVG(polling_places) as av_polls,
           STDDEV(polling_places) as stddev_polls,
           AVG(future_schools) AS avg_future,
           STDDEV(future_schools) AS stddev_future
        FROM FINALDATA
)

INSERT INTO ZScores(sa2_code, sa2_name, retail_zs, health_zs, primary_zs, secondary_zs, future_zs, stops_zs, polling_zs, geom)
Select D.sa2_code, D.sa2_name, (D.r_per_1000 - S.avg_retail)/S.stddev_retail, (D.h_per_1000 - S.avg_health)/ S.stddev_health,
    (D.primary_schools - S.avg_prim)/S.stddev_prim, (D.secondary_schools - S.avg_second)/S.stddev_second, 
    (D.future_schools - S.avg_future)/S.stddev_future, (D.tansport_count - S.av_stops)/S.stddev_stops, 
    (D.polling_places - S.av_polls)/S.stddev_polls, D.geom
FROM FinalData AS D, stats as S
"""

query(conn, sql)
```

## 4.1 Webscraping Crimes Dataset, Converting to SQL 

```python
webpage_source = requests.get("https://sydneysuburbreviews.com/total-crime-stats/").text
content = BeautifulSoup(webpage_source, 'html5lib')
web_table_headings = content.find('tr', 'row-1 odd')
```

```python
headings = []
for heading in web_table_headings.find_all('th'):
    text = heading.text.lower() 
    text = "_".join(text.split(" ")) 
    headings.append(text)

crime_stats = pd.DataFrame(columns=headings)
crime_stats.info()
```

```python
row_vals = content.find('tbody')

suburbs = []
crimes = []
average = []
crime_rates = []
for i in row_vals.find_all('tr'):
    s = i.find('td', 'column-1').text
    tc = i.find('td', 'column-2').text
    ai = i.find('td', 'column-3').text
    cr = i.find('td', 'column-4').text.strip("%")
    
    suburbs.append(s)
    crimes.append(tc)
    average.append(ai)
    crime_rates.append(cr)
    
crime_stats['suburb'] = suburbs
crime_stats['total_crimes'] = crimes
crime_stats['avg_incidents'] = average
crime_stats['crime_rate_per_capita'] = crime_rates
```

## 4.3 Join Tables, Calculate ZScores

```python
sql = """
DROP TABLE IF EXISTS CrimesDat;
CREATE TABLE CrimesDat AS
SELECT F.*, SUM(CAST(C.total_crimes AS FLOAT))/1000 AS crimes_per_1000
    FROM FinalData as F
    JOIN Crimes AS C
    ON F.sa2_name LIKE CONCAT('%', C.suburb, '%')
    GROUP BY F.sa2_code, F.sa2_name, F.geom, F.r_per_1000, F.h_per_1000, F.primary_schools, F.secondary_schools, F.future_schools, F.tansport_count, F.polling_places
"""

query(conn, sqlalchemy.text(sql))
```

```python
query(conn, "select * from crimesdat")
```

```python
sql = """
DROP TABLE IF EXISTS AddZScores;
CREATE TABLE AddZScores(
    sa2_code int,
    sa2_name VARCHAR(50),
    retail_zs float,
    health_zs float,
    primary_zs float,
    secondary_zs float,
    future_zs float,
    stops_zs float,
    polling_zs float,
    crimes_zs float,
    geom GEOMETRY(MULTIPOLYGON, 4326)
)
"""
query(conn, sql)
```

```python
sql = """
WITH stats AS(
    SELECT AVG(r_per_1000) AS avg_retail,
           STDDEV(r_per_1000) AS stddev_retail,
           AVG(h_per_1000) AS avg_health,
           STDDEV(h_per_1000) AS stddev_health,
           AVG(primary_schools) as avg_prim,
           STDDEV(primary_schools) as stddev_prim,
           AVG(secondary_schools) as avg_second,
           STDDEV(secondary_schools) as stddev_second,
           AVG(tansport_count) as av_stops,
           STDDEV(tansport_count) as stddev_stops,
           AVG(polling_places) as av_polls,
           STDDEV(polling_places) as stddev_polls,
           AVG(future_schools) AS avg_future,
           STDDEV(future_schools) AS stddev_future,
           AVG(crimes_per_1000) AS avg_crimes,
           STDDEV(crimes_per_1000) AS stddev_crimes
        FROM CrimesDat
)

INSERT INTO AddZScores(sa2_code, sa2_name, retail_zs, health_zs, primary_zs, secondary_zs, future_zs, stops_zs, polling_zs, crimes_zs, geom)
Select D.sa2_code, D.sa2_name, (D.r_per_1000 - S.avg_retail)/S.stddev_retail, (D.h_per_1000 - S.avg_health)/ S.stddev_health,
    (D.primary_schools - S.avg_prim)/S.stddev_prim, (D.secondary_schools - S.avg_second)/S.stddev_second, 
    (D.future_schools - S.avg_future)/S.stddev_future, (D.tansport_count - S.av_stops)/S.stddev_stops, 
    (D.polling_places - S.av_polls)/S.stddev_polls, (D.crimes_per_1000 - S.avg_crimes)/S.stddev_crimes, D.geom
FROM CrimesDat AS D, stats as S
"""

query(conn, sql)
```

```python
sql = """
ALTER TABLE AddZScores
ADD COLUMN sigmoid float;

UPDATE AddZScores
SET sigmoid = retail_zs + health_zs + (primary_zs + secondary_zs + future_zs)/3 + stops_zs + polling_zs - crimes_zs;
"""
query(conn, sql)
```

```python
sql = """
DROP TABLE IF EXISTS AllData;
CREATE TABLE AllData AS 
    SELECT C.*, COUNT(R.type) AS total_recreation
        FROM CrimesDat AS C
        RIGHT JOIN Recreation AS R
        ON ST_Intersects(C.geom, R.geom)
        GROUP BY C.sa2_code, C.sa2_name, C.geom, C.r_per_1000, C.h_per_1000, C.primary_schools, 
        C.secondary_schools, C.future_schools, C.tansport_count, C.polling_places, C.crimes_per_1000
"""
query(conn, sql)
```

```python
sql = """
WITH stats AS(
    SELECT AVG(r_per_1000) AS avg_retail,
           STDDEV(r_per_1000) AS stddev_retail,
           AVG(h_per_1000) AS avg_health,
           STDDEV(h_per_1000) AS stddev_health,
           AVG(primary_schools) as avg_prim,
           STDDEV(primary_schools) as stddev_prim,
           AVG(secondary_schools) as avg_second,
           STDDEV(secondary_schools) as stddev_second,
           AVG(tansport_count) as av_stops,
           STDDEV(tansport_count) as stddev_stops,
           AVG(polling_places) as av_polls,
           STDDEV(polling_places) as stddev_polls,
           AVG(future_schools) AS avg_future,
           STDDEV(future_schools) AS stddev_future,
           AVG(crimes_per_1000) AS avg_crimes,
           STDDEV(crimes_per_1000) AS stddev_crimes,
           AVG(total_recreation) AS avg_recreation,
           STDDEV(total_recreation) AS stddev_recreation
        FROM AllData
)

INSERT INTO FinalZScores(sa2_code, sa2_name, retail_zs, health_zs, primary_zs, secondary_zs, future_zs, stops_zs, polling_zs, crimes_zs, recreation_zs, geom)
Select D.sa2_code, D.sa2_name, (D.r_per_1000 - S.avg_retail)/S.stddev_retail, (D.h_per_1000 - S.avg_health)/ S.stddev_health, 0,
    (D.secondary_schools - S.avg_second)/S.stddev_second, 
    (D.future_schools - S.avg_future)/S.stddev_future, (D.tansport_count - S.av_stops)/S.stddev_stops, 
    (D.polling_places - S.av_polls)/S.stddev_polls, (D.crimes_per_1000 - S.avg_crimes)/S.stddev_crimes, 
    (D.total_recreation - S.avg_recreation)/S.stddev_recreation, D.geom
FROM AllData AS D, stats as S
"""

query(conn, sql)
```
